```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Lab_1/Au_nanoparticle_dataset.csv')
df.head()
```

```python
processed_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
processed_df.head()
```

```python
figure, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
figure.suptitle('Histograms of each feature')

N_total = axes[0]
N_bulk = axes[1]
N_surface = axes[2]
R_avg = axes[3]

N_total.hist(df['N_total'], bins=100)
N_total.set_ylabel('Frequency')
N_total.set_title('N_total')

N_bulk.hist(df['N_bulk'], bins=100)
N_bulk.set_title('N_bulk')
N_bulk.set_title('N_total')

N_surface.hist(df['N_surface'], bins=100)
N_surface.set_title('N_surface')
N_surface.set_title('N_total')

R_avg.hist(df['R_avg'], bins=100)
R_avg.set_title('R_avg')
R_avg.set_title('N_total')

plt.show()

```
